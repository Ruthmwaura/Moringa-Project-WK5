# ANALYSIS

```python
# QUESTION 1: FINDING THE CHANGES IN MORTALITY RATE FROM WHEN THEREWAN NO VACCINE TO WHEN THERE WAS A VACCINE

# let's find the mortality when there were no vaccines at the time of the data

data.loc[data['total_vaccinations'] == 0, 'total_deaths'].head(5)

# let's find the maximum value of total_deaths

data.loc[data['total_deaths'] == data['total_deaths'].max(), 'total_deaths']

# let's find the mortality when there were an abundance of vaccines

data.loc[data['total_vaccinations'] > 0, 'total_deaths'].tail(5)
```

#### The inference for the investigation above has been documented below

```python
# In the data provided, certain regions such as Afghanistan had no data provided on mortality rates at the start
# so our inference is that the mortality rate was 0 then. We can see this in the results above. We can also see that 
# the mortality rised to a maximum of 5886957.0 in the data provided and then fell again when there was an abundance of vaccines.
```

```python
# QUESTION 2: FINDING THE AREAS WITH LOW VACCINE COVERAGE

# let's find the areas with low vaccine supply and print the first 10 rows

data['total_vaccinations'].groupby(data['location']).min().sort_values(axis=0, ascending=False).head(10)
```

```python
# QUESTION 4:

#Let’s find the overall change in population from the COVID-19 pandemic in:
# Kenya, Malaysia, Spain, Venezuela, Panama

data['population_diff'] = data['population'] - data['total_deaths']
data.tail()

country = data[(data['location'] == 'Kenya')]
change_kenya = (country['population_diff'].max() - country['population_diff'].min())
overall_change = (change_kenya / country['population_diff'].max() *100)

print("The overall change in population was", change_kenya, "and it by", overall_change, "percent")
```

#### The inference for the investigation above has been documented below

```python
# 5635 died in Kenya. This was a 0.01% change in population
```

```python
# QUESTION 6:

# Let’s investigate the changes in stringency index as a result of COVID deaths in Kenya, Malaysia, Spain, Venezuela, Panama 
# Find differences between first and last months in stringency index

# let's find the difference in stringency index for kenya

data.loc[data['location'] == 'Kenya', 'stringency_index'].max() - data.loc[data['location'] == 'Kenya', 'stringency_index'].head(1)

# let's find the difference in stringency index for Malaysia

data.loc[data['location'] == 'Malaysia', 'stringency_index'].max() - data.loc[data['location'] == 'Malaysia', 'stringency_index'].head(1)

# let's find the difference in stringency index for Spain

data.loc[data['location'] == 'Spain', 'stringency_index'].max() - data.loc[data['location'] == 'Spain', 'stringency_index'].head(1)

# let's find the difference in stringency index for Venezuela

data.loc[data['location'] == 'Venezuela', 'stringency_index'].max() - data.loc[data['location'] == 'Venezuela', 'stringency_index'].head(1)

# let's find the difference in stringency index for Panama

data.loc[data['location'] == 'Panama', 'stringency_index'].max() - data.loc[data['location'] == 'Panama', 'stringency_index'].head(1)
```

#### The inference for the investigation above has been documented below

```python
# From the investigation conducted above, we can see that the stringency index for Kenya, Malaysia, Spain, Venezuela, and Panama steadily increased from the first day of month 
# to an all time high when COVID-19 was at its peak. This is because governments were trying to be strict on its policies for COVID-19 safety measures.
```
